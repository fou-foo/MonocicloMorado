```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import CARTODBPOSITRON
from bokeh.palettes import Spectral6, Viridis256, YlGn9
from bokeh.models import ColumnDataSource, LabelSet, OpenHead

from bokeh.transform import linear_cmap

def fn_bokeh_estaciones(dataframe):

    output_notebook()

    p = figure(x_axis_type="mercator", y_axis_type="mercator")

    source = ColumnDataSource(data=dict(
                                x = list(dataframe['coords_latitude']), 
                                y = list(dataframe['coords_longitude']),
                                x_end = list(dataframe['coords_latitude_end']),
                                y_end = list(dataframe['coords_longitude_end']),
                                tot_viajes=list(dataframe['tot_viajes']),
                                nombres =list(dataframe['est_retiro']) ))

    labels = LabelSet(x='x', y='y', text='nombres', level='glyph',
                  x_offset=5, y_offset=5, source=source, render_mode='canvas')

    mapper = linear_cmap(field_name='tot_viajes', 
                         palette=Viridis256,
                         low=min(dataframe['tot_viajes']),
                         high=max(dataframe['tot_viajes']))

    p.add_tile(CARTODBPOSITRON)

    p.circle(x = 'x',
             y = 'y',
             source = source,
             size=12,
             color=mapper)

    from bokeh.models import Arrow, VeeHead

    arrows = Arrow(end=OpenHead(line_color="black", size=10,# fill_alpha=.5,
                                line_width=2,
                                line_alpha=.5),
                       x_start='x',
                       y_start='y',
                       x_end='x_end',
                       y_end='y_end',
                       source=source,
                       line_color=mapper,
                       line_width=3)
                       #line_alpha='a')

    p.add_layout(labels)
    p.add_layout(arrows)

    show(p)
```

```python
df = df_brief_1_5
df['lat_vect'] = df['lat_ret'] + df['lat_vct']*400
df['lon_vect'] = df['lon_ret'] + df['lon_vct']*400
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import CARTODBPOSITRON
from bokeh.palettes import Spectral6, Viridis256, YlGn9
from bokeh.models import ColumnDataSource, LabelSet

from bokeh.transform import linear_cmap

output_notebook()

p = figure(x_axis_type="mercator", y_axis_type="mercator")

source = ColumnDataSource(data=dict(
                            x = list(df['lat_ret']), 
                            y = list(df['lon_ret']),
                            x_end = list(df['lat_vect']),
                            y_end = list(df['lon_vect']),
                            tot_viajes=list(df['trafico_entrante']),
                            nombres =list(df['est_retiro']) ))

labels = LabelSet(x='x', y='y', text='nombres', level='glyph',
              x_offset=5, y_offset=5, source=source, render_mode='canvas')

mapper = linear_cmap(field_name='tot_viajes', 
                     palette=Viridis256,
                     low=min(df['trafico_entrante']),
                     high=max(df['trafico_entrante']))

p.add_tile(CARTODBPOSITRON)

p.circle(x = 'x',
         y = 'y',
         source = source,
         size=12,
         color=mapper)

from bokeh.models import Arrow, VeeHead

arrows = Arrow(end=VeeHead(fill_color="black", size=10, fill_alpha=.5, line_alpha=.5),
                   x_start='x',
                   y_start='y',
                   x_end='x_end',
                   y_end='y_end',
                   source=source,
                   line_color=mapper,
                   line_width=5,
                   line_alpha=0.8)

p.add_layout(labels)
p.add_layout(arrows)

show(p)
```
